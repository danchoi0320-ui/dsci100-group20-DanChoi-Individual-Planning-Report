```r
#Let's add visualization to help us understand the data
#Visualizing played hours in bar graph

options(repr.plot.width = 12, repr.plot.height = 6)

ggplot(player_data, aes(x = played_hours)) +
  geom_histogram(binwidth = 1, fill = "skyblue", color = "black", alpha = 0.7) +
  labs(
    title = "Distribution of Played Hours",
    x = "Played Hours",
    y = "Number of Players"
  ) +
  theme_minimal(base_size = 16)
```

The plot shows that most players plays less than 25 in total. These results indicate that it may be challenging for us to find engaged players who have played sessions exceeding 1 hour, as many players' total played hours are less than 1 hour, which is marked as 0 on the graph.

```r
#Visualing played hours for each session.
#First, wrangle the data to convert original_start_time	and original_end_time from UNIX time to seconds.
library(dplyr)
library(lubridate)

#Mutate start_time and end_time to only have time
session_wrangled <- session_data |>
  mutate(
    start_dt = as.POSIXct(start_time, format = "%d/%m/%Y %H:%M"),
    end_dt   = as.POSIXct(end_time,   format = "%d/%m/%Y %H:%M"),
    start_time = format(start_dt, "%H:%M"),
    end_time   = format(end_dt, "%H:%M"),

#Converting end_time and start_time to dbl with decimals
    decimal_start_hr = hour(start_dt) + minute(start_dt)/60,
    decimal_end_hr   = hour(end_dt)   + minute(end_dt)/60,

#Calculating duration of each session in minutes
    duration_min = as.numeric(difftime(end_dt, start_dt, units = "mins"))) |>
select(hashedEmail, start_time, end_time, decimal_start_hr, decimal_end_hr, duration_min)

session_wrangled
```

```r
#Now ploting graph to visualize how many sessions are over 60 minutes
ggplot(session_wrangled, aes(x = duration_min >= 60)) +
  geom_bar(fill = "pink", color = "black") +
  scale_x_discrete(labels = c("FALSE" = "Below 60 min", "TRUE" = "Over 60 min")) +
  labs(
    title = "Number of Sessions Over/Under 60 Minutes",
    x = "Session Length",
    y = "Number of Sessions"
  ) +
  theme_minimal(base_size = 16)
```

We can see from the graph that most sessions are indeed below 1 hour, as shown in the "Distribution of Played Hours" graph above. This suggests that most players are not engaged, as defined by our criteria. Hence, it will limit and lower the accuracy of our prediction using regression, which may cause problems in future.

### Methods and Plan

To answer question: "Which time windows during the day and which days of the week are most likely to have a large number of engaged players?", I will use regression. Regression is appropriate because the goal is to predict how the number of engaged players changes with time-related variables such as hour of day and day of week. It is assumed that number of players we will remain consistant by time and day of the week enough for us to predict highest number of players per time and day. This method will show weakness if number of players vary with no pattern such as day or time. I will compare models using cross-validation and metrics such as RMSE. The data will be split into training and testing sets by 70/30, with the split performed after wrangling and cleaning. Cross-validation will serve as the validation process.
